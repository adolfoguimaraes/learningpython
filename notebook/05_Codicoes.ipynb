Para que possamos resolver esse problema, precisamos estudar mais um comando nas linguagens de programação: as condicionais. 

Em outras palavras, as condicionais são usadas quando precisamos selecionar quando e qual parte do código deve ser executada.

Isso é um recurso fundamental nas linguagens de programação. É fácil perceber em um programa de computador, em um jogo, que determinadas ações só acontecem se uma outra ação ocorrer. "Esse se" só é permitido graças as condicionais (o comando `if`). 

### Vamos começar ... 

As condicionais são escritas em Python da seguinte forma: 

```
if condição:
    bloco se a condição for verdadeira
```

Entenda por `bloco se a condição for verdadeira`, um trecho de código que deve ser executado caso a condição colocada seja de fato verdadeira. 

Antes de resolvermos o problema apresentado no início da aula, vamos trabalhar com um exemplo mais simples. 

**Exemplo:** Leia dois números e imprima a mensagem `O primeiro número é maior que o segundo` se isso for verdadeiro. Os números devem ser lidos com o comando `input`.

````python
# Ler dois números do usuário

numero1 = float(input("Digite o número 1: "))
numero2 = float(input("Digite o número 2: "))

# Testa se o número 1 é maior que número 2

if numero1 > numero2:
    print("O primeiro número é maior que o segundo")
````

Observe que a mensagem só é impressa quando o `numero1` é maior que o `numero2`. Isso acontece porque o trecho de código que exibe a mensagem está dentdo de um bloco separado pela condicional `if numero1 > numero2`. Esse trecho de código só é executado quando a condição é verdadeira.

Vamos aumentar o nosso problema. 

Além de imprimir a mensagem `O primeiro número é maior que o segundo`, vamos multiplicar o `numero2` por 5 quando isso acontecer. Ao final do programa deve-se imprimir os valores de `numero1` e `numero2`.

````python
# Ler dois números do usuário

numero1 = float(input("Digite o número 1: "))
numero2 = float(input("Digite o número 2: "))

# Testa se o número 1 é maior que número 2

if numero1 > numero2:
    print("O primeiro número é maior que o segundo")
    numero2 = numero2 * 5
    
# Imprime a mensagem final 

print("Os valores de numero1 e numero2 são %.1f e %.1f, respectivamente" % (numero1, numero2))
````

### E se quisermos imprimir uma mensagem quando o aluno não for aprovado? 

Neste caso, podemos usar um outro `if` para a condição correspondente. Vamos ver como fica:

````python
# Ler a nota 1 do aluno 

nota1 = float(input("Digite a nota 1 do aluno: "))

# Ler a nota 2 do aluno

nota2 = float(input("Digite a nota 2 do aluno: "))

# Calcula a média do aluno 

media = (5*nota1 + 7*nota2) / 12

print("O aluno tirou %.1f na primeira unidade e %.1f na segunda unidade. A média final foi de %.1f."
    % (nota1, nota2, media))

if media >= 6:
    print("O aluno foi aprovado na disciplina." )
    
if media < 6:
    print("O aluno foi reprovado na disciplina")
````

Desta forma conseguimos direcionar a execução do nosso programa para a parte do código que faz sentido de acordo com o que queremos. Observe que no código anterior, a `linha 17` só é executada quando a `media >= 6` e a `linha 20` quando a `media < 6`.

### Podemos melhorar nossas condicionais

Vamos voltar ao exemplo da média::

````python
# Ler a nota 1 do aluno 

nota1 = float(input("Digite a nota 1 do aluno: "))

# Ler a nota 2 do aluno

nota2 = float(input("Digite a nota 2 do aluno: "))

# Calcula a média do aluno 

media = (5*nota1 + 7*nota2) / 12

print("O aluno tirou %.1f na primeira unidade e %.1f na segunda unidade. A média final foi de %.1f."
    % (nota1, nota2, media))

if media >= 6:
    print("O aluno foi aprovado na disciplina." )
    
if media < 6:
    print("O aluno foi reprovado na disciplina")
````

Observe que nesse exemplo, não existe um valor no qual as duas mensagens seriam impressas. Nesse caso, podemmos substituir o segundo `if` pelo `else`. O `else` indica que se a condicional do `if` for falsa, o trecho do código após o `else` será executado. 

````python
# Ler a nota 1 do aluno 

nota1 = float(input("Digite a nota 1 do aluno: "))

# Ler a nota 2 do aluno

nota2 = float(input("Digite a nota 2 do aluno: "))

# Calcula a média do aluno 

media = (5*nota1 + 7*nota2) / 12

print("O aluno tirou %.1f na primeira unidade e %.1f na segunda unidade. A média final foi de %.1f."
    % (nota1, nota2, media))

if media >= 6:
    print("O aluno foi aprovado na disciplina." )
else:
    print("O aluno foi reprovado na disciplina")
````

Observe que o uso do `else` deixa o comando de condicional mais simples. No entanto, lembre-se que ele deve ser usado quando uma condição anula a execução da outra. Em termos de execução, essa abordagem também é interessante porque evita uma comparação adicional. 

Por exemplo, do jeito que estava antes, mesmo a média sendo `>=6`, a condicional do `media < 6` seria executada. Com o `else` isso não acontece. 

### Nunca é pouco exercitar. Vamos para mais um exemplo ... 

Os funcionários de uma empresa foram agraciados no final do ano com um bônus de natal. Tal gratificação será paga na folha de pagamento e depende da faixa salarial de cada um. Para os funcionários que ganham até R\$ 1500, o bônus é de 15%, Aqueles que ganham entre R\$ 1500 e R\$ 2500, o bônus é de 12,5% e por fim, os funcionários que ganham acima de R\$ 2500 terão uma bonificação de 10%. Elabore um programa que leia o salário do funcionário e calcule seu novo salário acrescido da bonificação correspondente. Imprima na tela o salário anterior, quando foi acrescido (em R\$) e o novo salário. Utilize mensagens adequadas para imprimir a informação.

````python
# Ler o salário do funcionário

salario = float(input("Digite o salário do funcionário: "))

if salario <= 1500:
    acrescimo = 0.150
else:
    if salario > 1500 and salario <= 2500:
        acrescimo = 0.125
    else:
        acrescimo = 0.100

acrescimo_reais = acrescimo * salario
novo_salario = salario + acrescimo_reais

print("Salário anterior do funcionário: R$ %.2f" % salario)
print("Bonificação de Natal: R$ %.2f" % acrescimo_reais)
print("Novo salário: R$ %.2f" % novo_salario)
````

Observer que para esse exemplo precisamos utilizar mais de um `if`/`else` para que pudéssemos resolver o problema proposto. Quando isso acontece, chamaos de `estruturas de if/else aninhadas`. Podíamos ter resolvido o problema com vários `if` mas já vimos que não é a forma mais adequada. 

### Atenção para a indentação do código

O código anterior possuía vários `if\else` aninhados. A medida que essa quantidade vai aumentando, devemos ter cuidado com a indentação. A imagem a seguir ilustra alguns aspectos da indentação no código anterior. 

````python
# Desconsidere estas linhas para o estudo da disciplina
Image(filename='images/05_image3.png', width=900) 
````

O problema de aninhar `if/else` é que a depender da quantidade de aninhamentos seu código fica muito grande e de difícil entendimento. Para evitar isso, podemos utilizar o **else if** que é representado pelo comando `elif`. 

Vamos analisar o problema a seguir: 

Em uma determinda loja, o preço dos produtos são definidos de acordo com a categoria. A tabela a seguir exibe as categorias e os preços correspondentes para cada categoria. 

| Categoria | Preço |
|---|---|
| 1 | 10,00 |
| 2 | 18,00 |
| 3 | 23,00 | 
| 4 | 26,00 |
| 5 | 31,00 |

Vamos ver como fica o programa usando somente `if` e `else`:

````python
# Exemplo retirado do livro do Nilo Menezes citado no início desse material

categoria = int(input("Digite a categoria do produto: "))

if categoria == 1:
    preco = 19
else:
    if categoria == 2:
        preco = 18
    else:
        if categoria == 3:
            preco = 23
        else:
            if categoria == 4:
                preco = 26
            else:
                if categoria == 5:
                    preco = 31
                else:
                    print("Categoria inválida, digite um valor entre 1 e 5!")
                    preco = 0

print("O preço do produto %i é: %.2f" % (categoria, preco) )
````

Perceba que muitos `if` e `else` aninhados dificulta a construção do programa, principalmente por conta do cuidado que devemos ter com a indentação. Uma forma de resolver isso é utilizar o comando `elif`. O `elif` substitui um par `else if`. Além da substituição, o `elif` cria um outro nível de estrutura sem criar deslocamentos à direita desnecessários. 

Vamos reescrever o código anterior utilizando o `elif`:

````python
# Exemplo retirado do livro do Nilo Menezes citado no início desse material

categoria = int(input("Digite a categoria do produto: "))

if categoria == 1:
    preco = 19
elif categoria == 2:
    preco = 18
elif categoria == 3:
    preco = 23
elif categoria == 4:
    preco = 26
elif categoria == 5:
    preco = 31
else:
    print("Categoria inválida, digite um valor entre 1 e 5!")
    preco = 0

print("O preço do produto %i é: %.2f" % (categoria, preco) )
````

Perceba as vantagens em se utilizar o `elif`. O código ficou menor. Além disso, o nível de aninhamentos foi reduzindo facilitando a leitura e entendimento do código construído.
