### Uma dúvida: como verificamos qual o resultado amazenado na variável *total*? 

Para que a gente possa exibir o valor de uma variável, devemos imprimir seu valor. Para isso, vamos utilizar o comando `print`. Considere o código a seguir:

```python
nota1 = 10
nota2 = 20

total = nota1 + nota2

print(total)
```

Simples assim? Sim, simples assim! 

A função `print` é uma função pré-definida do python que serve para exibir uma mensagem na tela. Essa mensagem pode ser um valor de uma variável ou uma expressão mais completa que vamos definir a seguir. No momento, basta saber que ela deve ser escrita sempre com `print` em minúsculo, seguido de dois parênteses `()`. Tudo aquilo que deseja-se imprimir deve ser colocado entre os parênteses. 

Vamos melhorar nossa mensagem: 

```python
print("O total é %i" % total)
```

A expressão `"O total é %i " % total"` colocada entre os parêntes é uma das formas de formatar o texto impresso. No final desta aula vamos ver várias maneiras de formatar um texto em Python. Por enquanto entenda que `%i` indica a posição que o número **i**nteiro vai ser impresso e `% total` indica que `%i` vai ser substituído pelo valor armazenado em `total`. 

### Outra dúvida: e se eu quiser digitar o valor de cada variável? 

Para isso podemos usar uma outra função pré-definida em Python: a função `input`. Esta função permite escrever uma mensagem na tela socilitando que o usuário digite alguma informação. Ela é escrita sempre com `input` em letra minúscula, seguido de dois parênteses. Entre os parênteses deve ser colocado o texto que vai ser impresso no momento da pergunta. 

No exemplo a seguir, vamos solicitar que usuário entre com duas notas (`nota1` e `nota2`) e no final imprimir o valor da soma destas duas notas na tela. 

```python
nota1 = input("Digite a nota 1: ")
nota2 = input("Digite a nota 2: ")

total = int(nota1) + int(nota2)

print("O total é %s " % total)
```

Vamos analisar o código anterior, linha por linha: 

* **Linha 1:** A função input vai imprimir a mensagem entre parênteses e aguardar que o usuário digite algum valor. Esse valor vai ser atribuído a variável `nota1`. Em outras palavras, o valor digitado pelo usuário vai ser armazenado no espaço de memória reservado para tal. 


* **Linha 2:** A linha dois faz quase a mesma coisa da linha 1. No entanto, a mensagem é diferente e o valor é atribuído a variável `nota2`. 

Até aqui, nada muito diferente do que a gente já tinha comentado. No entanto, a linha 4 tem um pequena diferença no momento que acessamos o valor de `nota1` e `nota2`. A explicação completa disso é dada na próxima seção, mas podemos já adiantar algumas coisas. 

Quando o usuário entra com as informações pelo comando `input`, esse valor é armazenado na memória como se fosse um texto. Aqui os valores em formato de texto são conhecidos como `string`. Quando tentamos somar esse valor com o `+`, o python não vai executar de forma correta porque ele espera, nesse caso, que os valores somados sejam **números inteiros**. Por isso usamos a função pré-definida `int` para transformar o valor de texto para inteiro. Devemos fazer isso, com cada valor lido. 

Sendo assim: 

* **Linha 4:** Transforma os valores de texto lidos em inteiros e realiza a soma. O resultado é armazenado na variável `total`. 

Por fim: 

* **Linha 5:** Imprime o valor armazenado na variável total formatado.
